```python
# date, ticker, v_y, oneminvol_y, count, day_ratio_15, lo_breakout, si_ratio_average7
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import tensorflow as tf
import glob

for csv in glob.glob("/content/drive/MyDrive/Barchart 2018 1.2 splitted/*.csv"):
  print('Working path:', csv)
  print('Working on:', csv.split('/')[-1])

  df = pd.read_csv(csv, low_memory= False)
  # df = pd.read_csv('/content/drive/MyDrive/Barchart 2018 Version 1.2.csv', low_memory= False)
  print(len(df))
  df.dropna(inplace=True)
  print(len(df))
  if len(df) == 0:
    print('Skipping, nothing to analyze...')
    continue
  for idx in range(1, 15):
    #########
    

    ##########
    X = df['v_y']
    input_list = ['oneminvol_y','count',
                'day_ratio_15', 'lo_breakout']
    for i, ele in enumerate(input_list):
      X = pd.concat([X, df[ele]], axis = 1)

    for ele in X.columns:
      X[ele] = X[ele].astype('float64')



    print('Running:', idx)
    y = df['change'+str(idx)]

    

    scaler = MinMaxScaler()
    X = scaler.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

    X_train = np.expand_dims(X_train, axis = 1)
    X_test = np.expand_dims(X_test, axis = 1)

    print("Input train shape:", X_train.shape)
    print("Output train shape:", y_train.shape)

    print("Input test shape:", X_test.shape)
    print("Output test shape:", y_test.shape)

    model = Sequential()
    model.add(LSTM(64, dropout=0.3, recurrent_dropout=0.3))
    # model.add(Dense(30, activation = 'relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                  optimizer='adam',
                  metrics=['accuracy'])
    
    model.fit(X_train, y_train,
            batch_size=100000,
            epochs=50,
            validation_data=(X_test, y_test))
    print('Evaluating >>>>>', idx)
    score, acc = model.evaluate(X_test, y_test,
                              batch_size=10000)  
    
    X = np.expand_dims(X, axis = 1)
    # X = np.squeeze(X, axis = 1)

    ## Predicting whole dataset:
    hist = model.predict(X)
    # hist = np.squeeze(hist, axis = 1)
    df['Predictions_'+'change'+str(idx)] = 0
    df['Predictions_'+'change'+str(idx)] = hist

  csv_name = '/content/drive/MyDrive/fintech/company_wise_prediction_2018/' + 'predictions_' + csv.split('/')[-1] 
  df.to_csv(csv_name, index = False)

    #$########
    ## Predictions only
    # X_try = np.expand_dims(X_try, axis = 1)
    # # X = np.squeeze(X, axis = 1)
    # hist = model.predict(X_try)
    # df_test['Predictions_'+'change'+str(idx)] = 0
    # df_test['Predictions_'+'change'+str(idx)] = hist
    ##########
```
